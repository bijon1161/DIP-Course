```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
img = cv2.imread('./image/flemingo.jpg')
```

**Histogram**

```python
#histogram
h_img = plt.hist(img.ravel(),256,[0,300])
plt.show()
```

```python
#histogram, equalization
gray_img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
he_img=cv2.equalizeHist(gray_img) 

plt.subplot(121)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.subplot(122)
plt.imshow(cv2.cvtColor(he_img, cv2.COLOR_BGR2RGB))
plt.title('histogtram equalized'); 
plt.show()
```

```python
row, col = 1, 2
fig, axs = plt.subplots(row, col, figsize=(15, 10))
fig.tight_layout()

axs[0].hist(img.ravel(),256,[0,256])
axs[0].set_title('Histogram')

axs[1].imshow(cv2.cvtColor(he_img, cv2.COLOR_BGR2RGB))
axs[1].set_title('Histogram Equalization')



plt.show()
```

**Morphological Operation**

```python
img1=cv2.imread('./image/j.jpg')
```

```python
#erosion
kernel=np.ones((5,5),dtype='uint8')

img_eroded=cv2.erode(img1,kernel,iterations=1)

#cv2.imshow('raw image',img1)
#cv2.imshow('img eroded',img_eroded)
#cv2.waitKey(0)
#cv2.destroyAllWindows()
plt.subplot(121)
plt.imshow(cv2.cvtColor(img1, cv2.COLOR_BGR2RGB))
plt.title('Raw Image')
plt.subplot(122)
plt.imshow(cv2.cvtColor(img_eroded, cv2.COLOR_BGR2RGB))
plt.title('Eroded Image')
plt.show()
```

```python
#dilation
img_dilated=cv2.dilate(img1,kernel,iterations=1)

#cv2.imshow('raw image',img1)
#cv2.imshow('img eroded',img_eroded)
#cv2.waitKey(0)
#cv2.destroyAllWindows()
plt.subplot(121)
plt.imshow(cv2.cvtColor(img1, cv2.COLOR_BGR2RGB))
plt.title('Raw Image')
plt.subplot(122)
plt.imshow(cv2.cvtColor(img_dilated, cv2.COLOR_BGR2RGB))
plt.title('Dilated Image')
plt.show()
```

```python
#opening

img_opening=cv2.morphologyEx(img1,cv2.MORPH_OPEN,kernel)

#cv2.imshow('raw image',img1)
#cv2.imshow('img eroded',img_eroded)
#cv2.waitKey(0)
#cv2.destroyAllWindows()
plt.subplot(121)
plt.imshow(cv2.cvtColor(img1, cv2.COLOR_BGR2RGB))
plt.title('Raw Image')
plt.subplot(122)
plt.imshow(cv2.cvtColor(img_opening, cv2.COLOR_BGR2RGB))
plt.title('After Opening Image')
plt.show()
```

```python
#closing

img_closing=cv2.morphologyEx(img1,cv2.MORPH_CLOSE,kernel)

#cv2.imshow('raw image',img1)
#cv2.imshow('img eroded',img_eroded)
#cv2.waitKey(0)
#cv2.destroyAllWindows()
plt.subplot(121)
plt.imshow(cv2.cvtColor(img1, cv2.COLOR_BGR2RGB))
plt.title('Raw Image')
plt.subplot(122)
plt.imshow(cv2.cvtColor(img_closing, cv2.COLOR_BGR2RGB))
plt.title('After Closing Image')
plt.show()
```

```python
row, col = 2, 2
fig, axs = plt.subplots(row, col, figsize=(15, 10))
fig.tight_layout()

axs[0][0].imshow(cv2.cvtColor(img_eroded, cv2.COLOR_BGR2RGB))
axs[0][0].set_title('Eroded Image')

axs[0][1].imshow(cv2.cvtColor(img_dilated, cv2.COLOR_BGR2RGB))
axs[0][1].set_title('Dilated Image')
 
axs[1][0].imshow(cv2.cvtColor(img_opening, cv2.COLOR_BGR2RGB))
axs[1][0].set_title('Image after Opening')

axs[1][1].imshow(cv2.cvtColor(img_closing, cv2.COLOR_BGR2RGB))
axs[1][1].set_title('Image after Closing')

plt.show()
```

**Filtering**

```python
img2 = cv2.imread('./image/bird.jpg')
#average blur

blur_img3=cv2.blur(img2,(9,9))

#gaussian blur

blur_gauss_img3=cv2.GaussianBlur(img2,(9,9),0)


#median blur(salt-pepper noise reduction)

blur_median_img3=cv2.medianBlur(img2,25)

row, col = 2, 2
fig, axs = plt.subplots(row, col, figsize=(15, 10))
fig.tight_layout()

axs[0][0].imshow(cv2.cvtColor(img2, cv2.COLOR_BGR2RGB))
axs[0][0].set_title('input image')

axs[0][1].imshow(cv2.cvtColor(blur_img3, cv2.COLOR_BGR2RGB))
axs[0][1].set_title('average blur')
 
axs[1][0].imshow(cv2.cvtColor(blur_gauss_img3, cv2.COLOR_BGR2RGB))
axs[1][0].set_title('gaussian blur')

axs[1][1].imshow(cv2.cvtColor(blur_median_img3, cv2.COLOR_BGR2RGB))
axs[1][1].set_title('median blur')

plt.show()
```
